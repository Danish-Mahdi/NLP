# Regex Functions


## 1. re.findall()
Return all non-overlapping matches of pattern in string, as a list of strings. The string is scanned left-to-right, and matches are returned in the order found.

```python
import re
string='Hello, Danish here!'
pattern= r'Danish here!'
re.findall(pattern, string)
```

```python
string='Hello, Danish3454 here! 8392382'
regex= '\d+'
re.findall(regex, string)
```

## 2. re.split()

```python
pattern='\W+'
string='Hello, How are u?'
re.split(pattern,string,maxsplit = 0)
# maxsplit = 0 means not specified
```

## 3. re.search()
This method either returns None (if the pattern doesn’t match), or a re.MatchObject contains information about the matching part of the string. This method stops after the first match, so this is best suited for testing a regular expression more than extracting data.

```python
import re 
regex = r"([a-zA-Z]+) (\d+)"

match = re.search(regex, "He wakes up at 8:00 am") 
if match != None:  
	print ("Time: %s" % (match.group(2))) 

else: 
	print ("The regex pattern does not match.") 

#     here r in regex pattern is for raw string. This way, Python knows not to process the backslashes in the string, passing the exact characters \ to the regex engine.
```

## 4. re.match()
This function in Python is used to determine if a regular expression pattern matches at the beginning of a string. It checks for a match only at the start of the string and returns a match object if a match is found; otherwise, it returns None.

```python
pattern = r"hello"
string1 = "hello world"
string2 = "say hello"

# Using re.match to find a match at the start of the string
match1 = re.match(pattern, string1)
match2 = re.match(pattern, string2)

if match1:
    print("Match found:", match1.group())
else:
    print("No match found in string1")

if match2:
    print("Match found:", match2.group())
else:
    print("No match found in string2")
```

# Difference between re.search() and re.match()

```python
re.match('cd', 'abcde')
re.search('cd', 'abcde')
```

#  Charting Word Length with NLTK

```python
from matplotlib import pyplot as plt
from nltk.tokenize import word_tokenize
sent='I am Danish I am learning Data Science'
tokens=word_tokenize(sent)
lengths=[len(w) for w in tokens]
plt.hist(lengths)
plt.show()
   
```
