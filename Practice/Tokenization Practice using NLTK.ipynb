### 2. Importing NLTK and Downloading Punkt
Here we import NLTK, the Natural Language Toolkit library, and download `punkt`. Punkt is a pre-trained tokenizer model in NLTK, designed for splitting text into sentences or words.

```python
import nltk
nltk.download('punkt')
# Here we imported NLTK: Natural Language Toolkit Library and Punkt. punkt is a pre-trained tokenizer model from NLTK, used for breaking text into sentences or words. It's language-independent and widely used for accurate text tokenization in NLP tasks. By downloading punkt, you gain access to robust tools for preprocessing text, making it easier to work with natural language data.
```

### 3. Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize 
text="I am Danish Mahdi from MUET JAMSHORO and I am a dedicated student. I am from 21 CS"
tokens=sent_tokenize(text)
print(tokens)
```

```python
from nltk.tokenize import sent_tokenize 
text = "مان ڊينش مھدي آهيان، ۽ مان هڪ محنتي شاگرد آهيان. مان 21 سي ايس جو آهيان."
tokens=sent_tokenize(text)
print(tokens)
```

### 4. Word Tokenization

##### i. Word Tokenization using word_tokenize

```python
from nltk.tokenize import word_tokenize 
text="I'm Danish Mahdi from 21CS MUET JAMSHORO, and I'm a dedicated student.I don't take interest in sports"
tokens=word_tokenize(text)
print(tokens)
```

##### ii. Word Tokenization using TreebankWordTokenizer. 'word_tokenize' is based on TreebankWordTokenizer

```python
from nltk.tokenize import TreebankWordTokenizer
text="I'm Danish Mahdi from 21CS MUET JAMSHORO, and I'm a dedicated student.I don't take interest in sports. "
tokenizer=TreebankWordTokenizer()
tokens=tokenizer.tokenize(text)
print(tokens)
```

##### iii. Word tokenization using punctuation. It tokenizes the text using punctuation and white spaces

```python
from nltk.tokenize import wordpunct_tokenize
text="I'm Danish Mahdi from 21CS MUET JAMSHORO, and I'm a dedicated student.I don't take interest in sports"
tokens=wordpunct_tokenize(text)
print(tokens)
```

##### iv. Word Tokenization using Regular Expression

```python
from nltk.tokenize import RegexpTokenizer
 
tokenizer = RegexpTokenizer(r'\w+')
text="I'm Danish Mahdi from 21CS MUET JAMSHORO, and I'm a dedicated student.I don't take interest in sports"
print(tokenizer.tokenize(text))
```
