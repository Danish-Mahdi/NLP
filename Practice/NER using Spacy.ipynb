```python
import spacy
nlp=spacy.load('en_core_web_sm')
# TRAINED MODEL FOR NER in ENG
```

```python
nlp.pipe_names
```

```python
doc=nlp("Muet Jamshoro is my university.I am originally from the Madeji city. ")
for ent in doc.ents:
    print(ent.text, '|', ent.label_, '|', spacy.explain(ent.label_))
```

```python
from spacy import displacy
displacy.render(doc,style='ent')
```

## Custom entities

```python
doc=nlp("Muet Jamshoro is my university.I am originally from the Madeji city. ")
for ent in doc.ents:
    print(ent.text, " | ", ent.label_)
```

```python
s = doc[11:12]
s
```

```python
from spacy.tokens import Span

s1 = Span(doc, 11,12, label='GPE')
doc.set_ents([s1],default='unmodified')
```

```python
for ent in doc.ents:
    print(ent.text, " | ", ent.label_)
```
